```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

Sigmoid - produces a S shaped curve mapping between 0 and 1.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

Tanh - produces smooth S shaped mapping between -1 and 1.

```python
def tanh(x):
    return np.tanh(x)
```

Relu - outputs the same input value if positive and zero if negative.

```python
def relu(x):
    return np.maximum(0, x)
```

Leaky relu - same as relu but allows a small, non zero gradient for -ve inputs.

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
```

Softmax - gives probability distribution values of vector scores that sum up to 1.

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)
```
